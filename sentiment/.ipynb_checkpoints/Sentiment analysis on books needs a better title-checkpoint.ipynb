Now, score the sentiment for individual piece using a labelled corpus. The R-package can be used to process the noisy signal you get. Okay, let's just jump right in!

You have to install a couple of packages to get this working, get [TextBlob](https://textblob.readthedocs.org/en/dev/), [afinn](https://github.com/fnielsen/afinn), and typical numpy, scipy stack.

```python
from textblob import TextBlob
from afinn import Afinn
import random
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = 20, 10
import scipy
from scipy.signal import spectrogram
import numpy as np
```

We will now define a function that takes a file name as input, splits it in to lines, and computes the sentiment for each line. The function will return the sentiment of each line as a python list.

```python
def sentiment(filename):
    with open (filename, "r") as myfile:
        text=myfile.read().replace('\n', ' ')
        blob = TextBlob(text)       
        tot = 0.0
        sent = []        
        afinn = Afinn()
        for sentence in blob.sentences:            
            senti = afinn.score(str(sentence))        
            sent.append(senti)   
        return sent
```

This function plots the sentiment by taking a file name and the title of the book.

```python
def sentiplot(filename, title):
    fig = plt.figure()
    sent= sentiment(filename)    
    plt.plot(sent,label=title)
    plt.ylabel("Cumulative Emotional Valence")
    plt.xlabel("Sentence #")    
    plt.legend()
```

That was it, just a few lines of code. We will pick up Romeo and Juliet as our first book.

So we have our sentiments for each sentence in the entire book. How to make sense of such a noisy signal? The first thing that came into my mind was a Fourier transform to see what frequencies exists in our signal. But, that would lose the entire information about the position. Is there a way to keep both the position and frequency informaton? Turns out that there is, it is called the short time fourier transform. 

```python
def pos_nes_chopper(s):
    p = s.copy()
    p[p<0] = 0
    n = s.copy()
    n[n>0] = 0
    return p,n
```

```python
# ugly code ahead
def stft_plot(pos,neg):
    plt.subplot(2,2,1)
    f,t,Sxx= spectrogram(pos,fs=1,nperseg=30,mode='magnitude')
    plt.pcolormesh(t, f, Sxx,shading='gouraud')
    plt.title('Positive Sentiment- Magnitude')
    plt.ylabel('Spatial Frequency')
    plt.xlabel('Sentence #')
    plt.colorbar()
    plt.subplot(2,2,2)
    f,t,Sxx= spectrogram(neg,fs=1,nperseg=30,mode='magnitude')
    plt.pcolormesh(t, f, Sxx,shading='gouraud')
    plt.title('Negative Sentiment - Magnitude')
    plt.ylabel('Spatial Frequency')
    plt.xlabel('Sentence #')
    plt.colorbar()
    plt.subplot(2,2,3)
    f,t,Sxx= spectrogram(pos,fs=1,nperseg=30,mode='psd')
    plt.pcolormesh(t, f, Sxx,shading='gouraud')
    plt.title('Positive Sentiment - Power Spectral Density')
    plt.ylabel('Spatial Frequency')
    plt.xlabel('Sentence #')
    plt.colorbar()
    plt.subplot(2,2,4)
    f,t,Sxx= spectrogram(neg,fs=1,nperseg=30,mode='psd')
    plt.pcolormesh(t, f, Sxx,shading='gouraud')
    plt.title('Negative Sentiment -  Power Spectral Denstiy')
    plt.ylabel('Spatial Frequency')
    plt.xlabel('Sentence #')
    plt.colorbar()   
```

```python
s = sentiment('pg1112.txt')
```

```python
pos, neg = pos_nes_chopper(np.array(s))
```

```python
# set color map
import colormaps as cmaps
plt.register_cmap(name='inferno', cmap=cmaps.inferno)
plt.set_cmap(cmaps.inferno)
```

```python
stft_plot(pos,neg)
```

Well, we can see that the first half of the novel does not have so much of negative sentiment, it is full of positive sentiment shown by the PSD and magnitude as well. Comparitively, the second half mostly has negative sentiment. This alligns nicely with my understanding of the tragic love story.
